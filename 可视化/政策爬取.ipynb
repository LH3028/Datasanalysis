```python
import requests
from bs4 import BeautifulSoup
import csv
headers={'User-Agent':'Request URL: http://www.fangchan.com/plus/nlist.php?tid=6&chinaarea=%E6%B5%99%E6%B1%9F&page=1'}
for i in range(1,2):
    link='http://www.fangchan.com/plus/nlist.php?tid=6&chinaarea=%E6%B5%99%E6%B1%9F&page'+str(i)
    r=requests.get(link,headers=headers)
    r.encoding='utf-8'
    soup=soup=BeautifulSoup(r.text,'lxml')
    print(r.text)
    house_list=soup.find_all('ul',class_="related-news-list")
    with open('news.csv', 'a',newline='',encoding='utf-8')as csvfile:
        w=csv.writer(csvfile)
        w.writerow(('标题','时间','URL'))
        for i in house_list[0].find_all("li"):
            temp=[]
            name=i.find('a').get_text()
            time=i.find('span').get_text()
            URL=i.find('a').get('href')
            temp=[name,time,URL]
            print(temp)
            w.writerow(temp)
     
```
